## SERIES

```python
#1.	Write a Pandas program to compare the elements of two Pandas series. 
#Sample Series [4,65,436,3,9], [7,0,3,897,9]


#First we declare the series and we compare them i.e see if they are the same then create a new dataframe to display
List_1 = pd.Series([4, 65, 436, 3, 9])
List_2 = pd.Series([7, 0, 3, 897, 9])

comparison = List_1 == List_2
df = pd.DataFrame({'List_1': List_1, 'List_2': List_2, 'Comparison': comparison})
df
```

```python
#2.	Write a Pandas program to add, subtract, multiply and divide two Pandas series. 
#Sample Series: [2,4,6,8,14], [1,3,5,7,9]

#we declare the series then perform the different operations on them
List_3 = pd.Series([2,4,6,8,14])
List_4 = pd.Series([1,3,5,7,9])

#Addition
Add = List_3 + List_4

#Subtraction
Subtract = List_3 - List_4

#Multiplication
Multiply = List_3 * List_4

#Division
Divide = List_3 / List_4

#display the results
df_1 = pd.DataFrame({'List_1': List_3, 'List_2': List_4, 'Addition': Add, 'Subtraction': Subtract, 'Multiplication': Multiply, 'Division': Divide})
df_1
```

```python
#5.Write a Pandas program to display the most frequent value in the given series 
#and replace everything else as ‘Other’ in the series.
#(Use the ‘HomeTeamGoals’ column in the dataset provided)

# Find the most frequent value
df_acon= pd.read_csv('AfricaCupOfNationsMatches.csv')
most_frequent_value = df_acon['HomeTeamGoals'].mode()[0]

# Replace all other values with 'Other'
df_acon['HomeTeamGoals'] = df_acon['HomeTeamGoals'].apply(lambda x: x if x == most_frequent_value else 'Other')

# Print the modified DataFrame
print(df_acon['HomeTeamGoals'])
```
